##What is Exploratory Data Analysis

Exploratory Data Analysis (EDA) is the process of exploring, investigating and gathering insights from data using statistical measures and visualizations. The objective of EDA is to develop and understanding of data, by uncovering trends, relationships and patterns.

EDA is both a science and an art. On the one hand it requires the knowledge of statistics, visualization techniques and data analysis tools like Numpy, Pandas, Seaborn etc. On the other hand, it requires asking interesting questions to guide the investigation and interpreting numbers & figures to generate useful insights.

# MATPLOTLIB CHARTS DESIGN

## Monthly Sales Trend

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["YearMonth"] = df["Order Date"].dt.to_period("M").astype(str)
monthly_revenue = df.groupby("YearMonth")["Price (INR)"].sum().reset_index()
plt.figure()
plt.plot(monthly_revenue["YearMonth"],monthly_revenue["Price (INR)"])
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Total Sales (INR)")
plt.title("Monthly Sales Trend")
plt.tight_layout()
plt.show()
```



## Insight: Sales Trend Over Time

- The line chart shows that sales fluctuate throughout the year.
- Overall, sales remain relatively stable between **6.5 – 6.8 million INR**.
- A noticeable drop in sales occurs in **February**.
- After February, sales gradually recover over the following months.
- Toward the end of the period, sales show **positive growth**.

## Monthly Revenue Trend

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["YearMonth"] = df["Order Date"].dt.to_period("M").astype(str)
monthly_revenue = df.groupby("YearMonth")["Price (INR)"].sum().reset_index()
plt.figure()
plt.bar(monthly_revenue["YearMonth"],monthly_revenue["Price (INR)"])
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Revenue (INR)")
plt.title("Monthly Revenue Trend")
plt.tight_layout()
plt.show()
```

## Insight: Monthly Revenue Trend (Jan – Aug 2025)

- The chart shows revenue variation from **January to August 2025**.
- Revenue reaches its **lowest point in February**.
- After February, revenue gradually **increases in the following months**.
- **May records the highest revenue**, indicating strong sales performance.
- Overall, revenue remains **stable between 6.5 and 6.8 million INR**.
- This trend demonstrates **consistent business growth**.

## Daily Sales Trend

```python
df["DayName"] = pd.to_datetime(df["Order Date"]).dt.day_name()
daily_revenue = (
    df.groupby("DayName")["Price (INR)"]
    .sum()
    .reindex(["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"])
)
plt.figure(figsize=(10,5))
plt.bar(daily_revenue.index, daily_revenue.values)
plt.title("Daily Revenue trend (Mon-Sun)")
plt.xlabel("Day")
plt.ylabel("Revenue (INR)")
plt.xticks(rotation=30)
plt.show()
```

### Insight: Daily Revenue Trend

- The Daily Revenue Trend chart shows that revenue remains relatively stable throughout the week.
- Revenue increases during the weekend.
- Saturday records the highest revenue, indicating higher customer demand for food delivery services.
- Tuesday has the lowest revenue, suggesting reduced weekday orders.
- Overall, weekends contribute significantly to the total revenue.

## Total Sales by Food Type(Veg vs Non-Veg)

```python
non_veg_keywords =[
    "chicken", "egg" , "fish", "mutton",
    "prawn" , "biryani" , "kabab", "kebab",
    "non-veg" , "non veg"
]
df["Food Category"] = np.where(
    df["Dish Name"].str.lower().str.contains("|".join(non_veg_keywords), na=False),
    "Non-Veg",
    "Veg"
)
```

```python
food_reveenue = (
    df.groupby("Food Category")["Price (INR)"]
    .sum()
    .reset_index()

)
```

```python
fig = px.pie(
    food_reveenue,
    values="Price (INR)",
    names="Food Category",
    hole=0.5,
    title="Reveenue Contribution: Veg vs Non-Veg",
)
fig.update_traces(
    textinfo="percent+label",
    pull=[0.05,0]
)
fig.update_layout(
    height=500,
    margin=dict(t=60,b=40,l=40,r=40)
)
fig.show()
```

## Insight: Revenue Contribution by Food Type

- The revenue contribution chart shows the share of revenue by food type.
- **Vegetarian dishes** contribute approximately **62.9%** of the total revenue.
- **Non-vegetarian dishes** contribute about **37.1%** of the total revenue.
- This indicates a **higher customer preference for vegetarian food items** on the platform.

## Total Sales by State

```python
  fig = px.bar(
      df.groupby("State",as_index=False)["Price (INR)"].sum()
      .sort_values("Price (INR)", ascending=False),
    x="Price (INR)",
    y="State",
    orientation="h",
    title="Revenue by State (INR)"
  )
fig.update_layout(height=600,yaxis=dict(autorange="reversed"))
fig.show()
```

## Quarterly Performance Summary

```python
df["Order_Date"] = pd.to_datetime(df["Order Date"])
df["Quarter"] = df["Order_Date"].dt.to_period("Q").astype(str)
quarterly_summary = (
    df.groupby("Quarter",as_index=False)
    .agg(
        Total_Sales=("Price (INR)","sum"),
        Avg_Rating=("Rating","mean"),
        Total_Orders=("Order_Date","count")
    )
    .sort_values("Quarter")
)
quarterly_summary
```

## Insight: Quarterly Performance Analysis

- The quarterly analysis shows variations in **revenue and order volume** throughout the year.
- **Q2 records the highest revenue and order volume**, indicating strong customer demand.
- In **Q3, both sales and orders decline** compared to the previous quarter.
- Despite these fluctuations, the **average customer rating remains stable**.
- The rating stays around **4.34**, suggesting **consistent service quality throughout the year**.

## Top 6 Cities by Sales

```python
top_6_cities =(
    df.groupby("City")["Price (INR)"]
        .sum()
        .nlargest(6)
    .sort_values()
       .reset_index()
)
fig = px.bar(
    top_6_cities,
    x="Price (INR)",
    y="City",
    title="Top 6 Cities by Sales (INR)",
    color_discrete_sequence=["pink"]
)
fig.show()
```

### Insight: City-wise Sales Analysis

- The city-wise sales analysis highlights the revenue contribution from different cities.
- **Bengaluru** generates the **highest revenue**, significantly outperforming other cities.
- **Lucknow, Hyderabad, and Mumbai** also contribute strongly with moderate sales levels.
- **New Delhi and Ahmedabad** show comparatively lower sales.
- These cities may present **opportunities for market expansion and promotional strategies**.

## Top 10 Food Categories by Average Price

```python
avg_price = df.groupby("Category")["Price (INR)"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=avg_price.values, y=avg_price.index)

plt.title("Top 10 Food Categories by Average Price")
plt.xlabel("Average Price (INR)")
plt.ylabel("Category")
color_discrete_sequence=["Blue"]
plt.show()
```

## Insight: Price Category Analysis

- **Chef’s Special Platters** and **Family Meal Boxes** fall into the **mid-range price category**, offering affordable options for families.
- Items like **Biryani Buckets** provide **comparatively lower-priced group meals**.
- Overall, **bulk and celebration meals dominate the higher price categories**.
